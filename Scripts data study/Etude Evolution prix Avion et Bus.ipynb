```python
df_bus_paris_stuttgart = pd.DataFrame(pd.read_csv('D:\Documents\ESILV\A5\WebScarpping\Bus\Stuttgart\\'+fichiers_stuttgart_bus[0]))
  
for i in range(1,len(fichiers_stuttgart_bus)):
    data = pd.read_csv('D:\Documents\ESILV\A5\WebScarpping\Bus\Stuttgart\\'+fichiers_stuttgart_bus[i])
    df = pd.DataFrame(data)
    df_bus_paris_stuttgart = pd.concat([df_bus_paris_stuttgart,df])
df_bus_paris_stuttgart.head()
```

# Transformation des données

```python
# On transforme la colonne des prix en int

for i in range(len(df_bus_bordeaux_lyon)):    
    df_bus_bordeaux_lyon['Prix_Train'][i] = df_bus_bordeaux_lyon['Prix_Train'][i][:-2]
    
df_bus_bordeaux_lyon['Prix_Train'] = pd.to_numeric(df_bus_bordeaux_lyon['Prix_Train']) 



for i in range(len(df_bus_paris_budapest)):
    df_bus_paris_budapest['Prix_Train'][i] = df_bus_paris_budapest['Prix_Train'][i][:-2]
    
df_bus_paris_budapest['Prix_Train'] = pd.to_numeric(df_bus_paris_budapest['Prix_Train']) 



for i in range(len(df_train_paris_lyon)):
    df_train_paris_lyon['Prix_Train'][i] = df_train_paris_lyon['Prix_Train'][i][:-2]
    
df_train_paris_lyon['Prix_Train'] = pd.to_numeric(df_train_paris_lyon['Prix_Train']) 



for i in range(len(df_train_paris_stuttgart)):
    df_train_paris_stuttgart['Prix_Train'][i] = df_train_paris_stuttgart['Prix_Train'][i][:-2]
    
df_train_paris_stuttgart['Prix_Train'] = pd.to_numeric(df_train_paris_stuttgart['Prix_Train'])



for i in range(len(df_bus_paris_lyon)):    
    df_bus_paris_lyon['Prix_Train'][i] = df_bus_paris_lyon['Prix_Train'][i][:-2]
    
df_bus_paris_lyon['Prix_Train'] = pd.to_numeric(df_bus_paris_lyon['Prix_Train']) 



for i in range(len(df_bus_paris_stuttgart)):    
    df_bus_paris_stuttgart['Prix_Train'][i] = df_bus_paris_stuttgart['Prix_Train'][i][:-2]
    
df_bus_paris_stuttgart['Prix_Train'] = pd.to_numeric(df_bus_paris_stuttgart['Prix_Train']) 
```
