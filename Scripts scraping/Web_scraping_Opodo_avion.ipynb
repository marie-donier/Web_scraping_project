```python
from selenium import webdriver
import time
import pandas as pd
from time import sleep
import datetime
```

```python
driver = webdriver.Chrome("C:/chrome_webdriver/chromedriver.exe")
#sleep(10)
```

```python
# 1-
web_link_t1 = 'https://www.opodo.fr/travel/#results/type=O;dep=2023-02-06;from=PAR;to=LYS;internalSearch=true'

# 2-
web_link_t2 = 'https://www.opodo.fr/travel/#results/type=O;dep=2023-02-06;from=PAR;to=STR;internalSearch=true'

# 3- 
web_link_t3 = 'https://www.opodo.fr/travel/#results/type=O;dep=2023-02-06;from=PAR;to=BUD;internalSearch=true'

# 4- 
web_link_t4 = 'https://www.opodo.fr/travel/#results/type=O;dep=2023-02-06;from=BOD;to=LYS;internalSearch=true'

#5-
web_link_t5 = 'https://www.opodo.fr/travel/#results/type=O;dep=2023-02-06;from=LYS;to=MRS;internalSearch=true'
```

```python
date_time = datetime.datetime.now()
result = str(date_time.day) + '-' + str(date_time.month) + '-' + str(date_time.year) + '-' + str(date_time.hour)

for i in range(0,5) :
    
    print(i)
    
    web_link = ''
    trajet = ''
    
    if i == 0 :
        web_link = web_link_t1
        trajet = 'Paris-Lyon'
    
    elif i == 1 :
        web_link = web_link_t2
        trajet = 'Paris-Stuttgart'
        
    elif i == 2 :
        web_link = web_link_t3
        trajet = 'Paris-Budapest'
        
    elif i == 3 :
        web_link = web_link_t4
        trajet = 'Bordeaux-Lyon'
        
    elif i == 4 :
        web_link = web_link_t5
        trajet = 'Lyon-Marseille'
        
    
    driver.get(web_link)
    sleep(10)
    
    
    if i == 0 :  # On ferme la page des cookies seulement pour la première requête, après elle n'apparait plus
        reject_cookies_button = driver.find_element_by_id('didomi-notice-disagree-button')
        #accept_cookies_button.text
        reject_cookies_button.click()
        
        
    driver.execute_script("window.scrollTo(0, 300)")
    sleep(2)
    driver.execute_script("window.scrollTo(0, 600)")
    sleep(2)
    driver.execute_script("window.scrollTo(0, 900)")
    sleep(2)
    driver.execute_script("window.scrollTo(0, 1200)")
    sleep(2)
    driver.execute_script("window.scrollTo(0, 1500)")
    sleep(2)
    driver.execute_script("window.scrollTo(0, 1800)")
    sleep(2)
    driver.execute_script("window.scrollTo(0, 2100)")
    sleep(2)
    driver.execute_script("window.scrollTo(0, 2400)")
    sleep(2)
    driver.execute_script("window.scrollTo(0, 2700)")
    sleep(2)
    driver.execute_script("window.scrollTo(0, 3000)")
    sleep(2)
    driver.execute_script("window.scrollTo(0, 3300)")
    sleep(2)
    driver.execute_script("window.scrollTo(0, 3600)")
    sleep(2)
    driver.execute_script("window.scrollTo(0, 3900)")
    sleep(2)
    driver.execute_script("window.scrollTo(0, 4200)")
    sleep(2)
    driver.execute_script("window.scrollTo(0, 4500)")
    sleep(2)
    driver.execute_script("window.scrollTo(0, 4800)")
    sleep(2)
    driver.execute_script("window.scrollTo(0, 5100)")
    sleep(2)
    driver.execute_script("window.scrollTo(0, 5400)")
    sleep(2)
    driver.execute_script("window.scrollTo(0, 5700)")
    sleep(2)
    driver.execute_script("window.scrollTo(0, 6000)")
    sleep(2)
    driver.execute_script("window.scrollTo(0, 6300)")
    sleep(2)
    driver.execute_script("window.scrollTo(0, 6600)")
    sleep(2)
    driver.execute_script("window.scrollTo(0, 6900)")
    sleep(2)
    driver.execute_script("window.scrollTo(0, 7200)")
    sleep(2)
    driver.execute_script("window.scrollTo(0, 7500)")
    sleep(2)
    driver.execute_script("window.scrollTo(0, 7800)")
    sleep(2)
    driver.execute_script("window.scrollTo(0, 8100)")
    sleep(2)
    driver.execute_script("window.scrollTo(0, 8400)")
    sleep(2)
    driver.execute_script("window.scrollTo(0, 8700)")
    sleep(2)
    driver.execute_script("window.scrollTo(0, 9000)")
    sleep(2)
    
    
    
    #RECUPERATION DES INFORMATIONS DES VOLS
    
    #Récupération date de départ
    date_depart = driver.find_element_by_xpath('//*[@id="depart_date_field_container"]/div/div[1]/div[1]/div[2]')
    date_depart = date_depart.text
    
    #Année de départ
    #annee_depart = ''
    #try :
    #    annee_depart = driver.find_element_by_xpath('//*[@id="depart_date_field_container"]//div[@class="odf-calendar-title"]')
    #    annee_depart = annee_depart.text[-4:len(annee_depart.text)]
    
    #except :
    #    annee_depart = driver.find_element_by_xpath('//*[@id="depart_date_field_container"]/div/div[1]/div[2]/div[1]/div[2]/div[1]/div[1]/div[1]')
    #    annee_depart = annee_depart.text[-4:len(annee_depart.text)]
    
    annee_depart = '2023'
    
   
    #Récupération nombre de passagers
    nb_passagers = driver.find_element_by_xpath('//*[@id="passengers_search_manager"]/div[2]/div/div[1]/div/span/div/span[2]')
    nb_passagers = nb_passagers.text
    
    # Nombre d'adultes
    nb_passagers_adulte = driver.find_element_by_xpath('//*[@id="passengers_search_manager"]/div[2]/div/div[2]/div[1]/div[1]/div/div/div[2]/div/div')
    nb_passagers_adulte = nb_passagers_adulte.text
    
    # Nombre d'enfants (2-11 ans)
    nb_passagers_enfant = driver.find_element_by_xpath('//*[@id="passengers_search_manager"]/div[2]/div/div[2]/div[1]/div[2]/div/div/div/div[2]/div/div')
    nb_passagers_enfant = nb_passagers_enfant.text
    
    # Nombre de bébés (> 2 ans)
    nb_passagers_bebe = driver.find_element_by_xpath('//*[@id="passengers_search_manager"]/div[2]/div/div[2]/div[1]/div[3]/div/div/div/div[2]/div/div')
    nb_passagers_bebe = nb_passagers_bebe.text
    
    #Récupération des heures des vols
    heures_depart_arrivee_aller_retour = driver.find_elements_by_xpath('//*[@id="results_list_container"]/div[1]/div/div[1]/div//div[@class="css-v0s8x5-BaseText-Body eezsimq0"]')
    
    aller_depart_heure = []
    aller_arrivee_heure = []

    for i in range(0, len(heures_depart_arrivee_aller_retour)-1, 2): 

        aller_depart_heure.append(heures_depart_arrivee_aller_retour[i].text)
        aller_arrivee_heure.append(heures_depart_arrivee_aller_retour[i+1].text)

        
    
    #Récupération des lieux de départ et d'arrivée des vols
    lieux_depart_arrivee_aller_retour = driver.find_elements_by_xpath('//*[@id="results_list_container"]/div[1]/div/div[1]/div//div[@class="css-13x7to7-BaseText-Body eezsimq0"]')
    
    aller_depart_aeroport = []
    aller_depart_ville = []
    aller_arrivee_aeroport = []
    aller_arrivee_ville = []


    for i in range(0, len(lieux_depart_arrivee_aller_retour)-3, 4): 

        aller_depart_aeroport.append(lieux_depart_arrivee_aller_retour[i].text)
        aller_depart_ville.append(lieux_depart_arrivee_aller_retour[i+1].text)
        aller_arrivee_aeroport.append(lieux_depart_arrivee_aller_retour[i+2].text)
        aller_arrivee_ville.append(lieux_depart_arrivee_aller_retour[i+3].text)

        
        
     # Récupération de la compagnie aérienne   
    compagnies_aller_retour = driver.find_elements_by_xpath('//*[@id="results_list_container"]/div[1]/div/div[1]/div//div[@class="css-1um4vyc-BaseText-Body eezsimq0"]')
    
    aller_compagnie = []

    for i in range(0, len(compagnies_aller_retour)): 

        aller_compagnie.append(compagnies_aller_retour[i].text)

        
    # Récupération de la durée du vol
    duree_vols_aller_retour = driver.find_elements_by_xpath('//*[@id="results_list_container"]/div[1]/div/div[1]/div//span[@class="css-1mbgvvh-BaseText-Text ek4n60o0"]')
    
    aller_duree = []

    for i in range(0, len(duree_vols_aller_retour)): 

        aller_duree.append(duree_vols_aller_retour[i].text)

    
    # Récupération du nombre d'escales
    nb_escales_aller_retour = driver.find_elements_by_xpath('//*[@id="results_list_container"]/div[1]/div/div[1]/div//span[@class="css-1kh4dda-BaseText-Text ek4n60o0"]')
    
    aller_escale = []

    for i in range(0, len(nb_escales_aller_retour)): 

        aller_escale.append(nb_escales_aller_retour[i].text)

        
    # Récupération du prix non réduit
    prix_non_reduit_aller_retour = driver.find_elements_by_xpath('//*[@id="results_list_container"]/div[1]/div/div[1]/div//span[@class="css-atmab-StyledContainer e1m61u850"]')
    
    prix_non_reduit = []

    for i in range(0, len(prix_non_reduit_aller_retour)): 

        prix_non_reduit.append(prix_non_reduit_aller_retour[i].text)
    
    
    
    # Récupération du prix réduit
    prix_reduit_aller_retour = driver.find_elements_by_xpath('//*[@id="results_list_container"]/div[1]/div/div[1]/div//div[@class="of_autos-itinerary-prime-price css-1x3sxzl-Box e17fzqxg0"]/span[@class="css-1iu0vlo-StyledContainer e16uabde3"]')
    
    prix_reduit = []

    for i in range(0, len(prix_reduit_aller_retour)): 

        prix_reduit.append(prix_reduit_aller_retour[i].text)

    
    # Type de bagage inclu -> pas utile pour notre étude finalement donc en commentaires
    #type_bagage_aller_retour = driver.find_elements_by_xpath('//*[@id="results_list_container"]/div[1]/div/div[1]/div//span[@class="css-1xh0qv8-BaseText-Text ek4n60o0"]')

    #aller_bagage = []

    #for i in range(0, len(type_bagage_aller_retour)): 

     #   aller_bagage.append(type_bagage_aller_retour[i].text)
        
    
    
    length = len(prix_non_reduit)
    length

    col_date_depart = []
    col_date_retour = []
    col_nb_passagers = []
    col_nb_passagers_adulte = []
    col_nb_passagers_enfant = []
    col_nb_passagers_bebe = []
    
    col_jour_releve = []
    col_mois_releve = []
    col_annee_releve = []
    col_heure_releve =[]


    for i in range(length):
        col_date_depart.append(date_depart)
        col_nb_passagers.append(nb_passagers)
        col_nb_passagers_adulte.append(nb_passagers_adulte)
        col_nb_passagers_enfant.append(nb_passagers_enfant)
        col_nb_passagers_bebe.append(nb_passagers_bebe)
        
        col_jour_releve.append(str(date_time.day))
        col_mois_releve.append(str(date_time.month))
        col_annee_releve.append(str(date_time.year))
        col_heure_releve.append(str(date_time.hour))
    
    
    # Construction du dataframe
    df = pd.DataFrame({"Ville_depart" : aller_depart_ville, "Ville_arrivee" : aller_arrivee_ville, "Date_depart": col_date_depart, "Nb_passagers": col_nb_passagers, "Horaire_depart": aller_depart_heure, "Horaire_arrivee": aller_arrivee_heure, "Aeroport_depart": aller_depart_aeroport, "Aeroport_arrivee": aller_arrivee_aeroport,  "Compagnie": aller_compagnie, "Duree": aller_duree, "Escale": aller_escale, "Prix": prix_non_reduit, "Releve_annee": col_annee_releve , "Releve_mois": col_mois_releve, "Releve_jour":  col_jour_releve, "Releve_heure": col_heure_releve})
    df.drop_duplicates(inplace = True)
    
    #On ajoute l'année aux dates de l'aller et du retour
    df["Date_depart"] = df["Date_depart"].str.slice(-5,len(df["Date_depart"][0])) + '/' + annee_depart
    
    # Export du dataframe sous forme d'un fichier csv
    df.to_csv("D:/ESILV/Année 5/Web Scraping/Projet/Avion/Scrap_data_OPODO_" + trajet + '_' + str(date_time.day) + '-' + str(date_time.month) + '-' + str(date_time.year) + '-' + str(date_time.hour) + '.csv', index = False)
    
    driver.execute_script("window.scrollTo(0, 0)")
    sleep(4)
    
```
